# Scrap IRIS Data from Wikipedia

## Import libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

## Soup Creation

```python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Iris_flower_data_set"

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

print(soup.title)
print(soup.title.text)

print(soup.prettify())  # This line prints the formatted HTML content

```

## For loop to find tables

```python
tables = soup.find_all("table")

dataframes = []  # List to store dataframes

for table in tables:
    rows = table.find_all('tr')[1:]  ## skip first row 
    data = []
    for row in rows:
        cols = row.find_all("td")
        cols = [col.text.strip() for col in cols]  # Call strip() method correctly
        data.append(cols)
    df = pd.DataFrame(data)
    dataframes.append(df)
```

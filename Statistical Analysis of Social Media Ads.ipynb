## Normality Check

### Age Column

\\ Null Hypothesis : Data is Normal

\\ Alternate Hypothesis : Data is not Normal

```python
sns.histplot(df['Age'], kde=1)
```

```python
sns.histplot(df["Age"], kde = True)
plt.axvline(df["Age"].mean(), color="Red")
plt.axvline(df["Age"].median(), color="Green")
plt.axvline(df["Age"].mode()[0], color="Blue")

# Print 
print("Mean : ", df["Age"].mean())
print("Median : ", df["Age"].median())
print("Mode : ", df["Age"].mode()[0])
```

```python
stats.shapiro(df['Age'])
```

##### Based on the results of the normality test conducted at a significance level of [pvalue=2.773022515611956e-06], we do not have sufficient evidence to accept the null hypothesis. Therefore, we do not conclude that the data significantly deviates from a normal distribution.

## Levene Test

```python
group1 = (df['Age'])
group2 = (df["EstimatedSalary"])

# Perform Levene Test
w_stat, p_val = stats.levene(group1, group2)

print("w_statistic:", w_stat)
print("p-value:", p_val)

if p_val < 0.05:  # Assuming alpha level of 0.05
    print(f"Variance are equal : Fail to Reject the null hypothesis")
else:
    print(f"Variance are not equal : Reject the null hypothesis")
```

##### Based on the results of the homoginity test conducted at a significance level of [p-value:p-value: 5.139400289012397e-22 , w_statistic: 98.73947769124322], (variance are  equal) we do not have sufficient evidence to reject the null hypothesis. Therefore, we do conclude that the data significantly deviates from a normal distribution.

### Chi-Squared Test

```python
contigency_table_1 = pd.crosstab(df["Gender"], df["Purchased"])
contigency_table_1
```

```python
chi2, p, dof, expected = stats.chi2_contingency(contigency_table_1)

print()
print("Chi_Square Statistic:", chi2)
print("p-value:", p)
print("Degrees of Freedom:", dof)
print("Expected:", expected)
```

### Man-Whittney Test (Independent Sampling)

```python
from scipy.stats import mannwhitneyu

group1 = df['Age']
group2 = df['EstimatedSalary']

# Perform Mann-Whitney U test
statistic, p_value = mannwhitneyu(group1, group2, alternative='two-sided')

print(f'Mann-Whitney U statistic: {statistic}, p-value: {p_value}')
```

#### Wilcoxon Test (Paired Sampling)

```python
from scipy.stats import wilcoxon

group1 = df['Age']
group2 = df['EstimatedSalary']

# Perform Mann-Whitney U test
statistic, p_value = wilcoxon(group1, group2, alternative='two-sided')

print(f'Mann-Whitney U statistic: {statistic}, p-value: {p_value}')
```

### Correlation

```python
def pearson(x, y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    x_std = np.std(x)
    y_std = np.std(y)
    n = len(x)
    return sum((x-x_mean)*(y-y_mean))/(n*x_std*y_std)

def spearman(x,y):
    x_rank = pd.Series(x).rank()
    y_rank = pd.Series(y).rank()
    return pearson(x_rank,y_rank)

# Example Dataset
x=(df["Age"])
y=(df["EstimatedSalary"])

print(f"Spearman Correlation Coefficient: {spearman(x,y)}")
# Differentiate all values
if spearman(x,y) < 0.6 > 0:
    print("Positive Relation")
elif spearman(x,y) > 0.6:
    print("Highly Positive Relation")
elif spearman(x,y) > -0.6 < 0:
    print("Negative Correlation")
elif  spearman(x,y) < -0.6:
    print("Highly Negative Correlation")
else:
    print("No Relation")
```

```python
data =pd.DataFrame({'Age': x, 'EstimatedSalary': y})
heatmap_data = data.corr(method='spearman')

plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Spearman Correlation Heatmap between Age and EstimatedSalary')
plt.show()
```

# Decision Tree

A decision tree is a supervised machine learning algorithm used for both classification and regression tasks. It's a flowchart-like structure where each internal node represents a "test" on an attribute (e.g., whether a feature is less than a certain value), each branch represents the outcome of the test, and each leaf node represents a class label or a numerical value. Decision trees are particularly popular because they are easy to understand and interpret.

***Where to use decision trees***:
1. **Classification**: Decision trees are commonly used for classification tasks where the goal is to predict the category or class label of a new instance. For example, classifying whether an email is spam or not spam.
2. **Regression**: Decision trees can also be used for regression tasks, where the goal is to predict a continuous value. For example, predicting house prices based on features such as area, number of bedrooms, etc.
3. **Feature Selection**: Decision trees can be used to identify the most important features in a dataset. Features that appear near the root of the tree are typically more important for predicting the target variable.
4. **Data Exploration**: Decision trees can help in understanding the underlying structure of the data by visualizing the decision-making process.

***Data requirements for decision trees***:
1. **Categorical and Numerical Data**: Decision trees can handle both categorical and numerical data types. They are versatile in this aspect.
2. **Preprocessed Data**: Data should be preprocessed, including handling missing values and encoding categorical variables into a numerical format if needed. Decision trees don't handle missing values internally, so they need to be addressed beforehand.
3. **Enough Data**: Decision trees can work well with small datasets, but they can also suffer from overfitting if the dataset is too small or noisy. Having enough data to represent the underlying structure of the problem is crucial.
4. **Balanced Data**: For classification tasks, having a balanced dataset (similar proportions of each class) is ideal. Imbalanced datasets might lead to biased models.
5. **Homogeneous Data**: Decision trees work best when the data is relatively homogeneous within each region defined by the splits. If there's a lot of variability within these regions, decision trees might not perform well.

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
```

## Import Dataset

```python
df = pd.read_csv('D:\\Data Practice JN\\Pre-Processing\\Wrangled Data of Social Media Ads.csv')
df
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
df.head()
```

## Model Building

### Define Model Values

```python
x = df[['Age', 'Gender']]
y = df[['Purchased']]
```

```python
x.head()
```

```python
y.head()
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
model = DecisionTreeClassifier()
model.fit(x_train , y_train)
```

## Normaity Check

### Age Column

\\ Null Hypothesis: Data is Normal.

\\ Alternate Hypothesis: Data is not Normal.

```python
sns.histplot(df['Age'], kde=1)
```

```python
sns.histplot(df["Age"], kde = True)
plt.axvline(df["Age"].mean(), color="Red")
plt.axvline(df["Age"].median(), color="Green")
plt.axvline(df["Age"].mode()[0], color="Blue")

# Print 
print("Mean : ", df["Age"].mean())
print("Median : ", df["Age"].median())
print("Mode : ", df["Age"].mode()[0])
```

```python
stats.shapiro(df['Age'])
```

##### Based on the results of the normality test conducted at a significance level of [pvalue=4.878829145127384e-07], we do not have sufficient evidence to accept the null hypothesis. Therefore, we do not conclude that the data significantly deviates from a normal distribution.

## Levene Test

\\ Null Hypothesis : Data is Normal.

\\ Alternate Hypothesis : Data is not Normal.

```python
group1 = (df['Age'])
group2 = (df["Na_to_K"])

# Perform Levene Test
w_stat, p_val = stats.levene(group1, group2)

print("w_statistic:", w_stat)
print("p-value:", p_val)

if p_val < 0.05:  # Assuming alpha level of 0.05
    print(f"Variance are equal : Fail to Reject the null hypothesis")
else:
    print(f"Variance are not equal : Reject the null hypothesis")
```

##### Based on the results of the homoginity test conducted at a significance level of [p-value: 0.8954460039049175, w_statistic: 0.01729232639401157], (variance are not equal) we do not have sufficient evidence to accept the null hypothesis. Therefore, we do not conclude that the data significantly deviates from a normal distribution.

### Correlation

```python
def pearson(x, y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    x_std = np.std(x)
    y_std = np.std(y)
    n = len(x)
    return sum((x-x_mean)*(y-y_mean))/(n*x_std*y_std)

def spearman(x,y):
    x_rank = pd.Series(x).rank()
    y_rank = pd.Series(y).rank()
    return pearson(x_rank,y_rank)

# Example Dataset
x=(df["Age"])
y=(df["Na_to_K"])

print(f"Spearman Correlation Coefficient: {spearman(x,y)}")
# Differentiate all values
if spearman(x,y) < 0.6 > 0:
    print("Positive Relation")
elif spearman(x,y) > 0.6:
    print("Highly Positive Relation")
elif spearman(x,y) > -0.6 < 0:
    print("Negative Correlation")
elif  spearman(x,y) < -0.6:
    print("Highly Negative Correlation")
else:
    print("No Relation")
```

```python
data =pd.DataFrame({'Age': x, 'Na_to_K': y})
heatmap_data = data.corr(method='spearman')

plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Spearman Correlation Heatmap between Age and Na_to_K')
plt.show()
```
